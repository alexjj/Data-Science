# Normalisation

Makes comparions between columns meaningful and computationally useful
e.g. linear regression large numbers will influence the data more, but might not be important.

normalise from 0-1

## Methods

### Simple feature scaling

new = old/max

Gives 0-1

### Min-Max

new = (old - min) / (max - min)

Gives 0-1

### Z-Score / Standard Score

new = (old - mean) / standard deviation

Range around zero and move between -3 nad 3

````python
# Use it on length:

# Simple feature scaling:

# df['length'] = df['length']/df['length'].max()

# min-max

# df['length'] = (df['length']-df['length'].min())/(df['length'].max()-df['length'].min())

# z-score

df['length'] = (df['length']-df['length'].mean())/df['length'].std()

df['length']
````

# Binning

Grouping of values into bins
Converts numerical to categorical
Could help reduce the number of values to have a better understanding of the distribution. Reduce the number of unique values

````python
# Bin the prices into low, medium and high

# make 3 bins so want 4 numbers as dividers to range all the bins
bins = np.linspace(min(df['price']), max(df['price']), 4)

# names of bins
group_names = ["low", "medium", "high"]

# Use cut to segment and put data into bins
df['price-binned'] = pd.cut(df['price'], bins, labels=group_names, include_lowest=True)

# Then may plot historgram to see where number of values in bins lie
````

# Grouping

Is any relationship between drive system and price, so then which drive system adds the most value to a vehicle. So want to group the same drive type.

Use groupby() method, with either single or multiple variables

````python
# avg price of vehicles and see how they differ between different types of body styles and drive wheels

# Make a smaller dataframe with just the columns we're interested in
df_test = df[['drive-wheels', 'body-style', 'price']]

# Group data and take mean of price
df_grp = df_test.groupby(['drive-wheels','body-style'], as_index=False).mean()
df_grp.sort_values(by=['price'], ascending=False)
````

````python
# Use pivot method to make it easier to visualise

df_pivot = df_grp.pivot(index='drive-wheels', columns='body-style')
df_pivot
````

````python
# Heatmap, plot target variable over multiple variables to get visual clues to relationships

sns.set(font_scale=1.2)
sns.set_style({"savefig.dpi": 100})
sns.heatmap(df_pivot, cmap=plt.cm.Blues, linewidths=.1)
````

# ANOVA - Analysis of variance

Find correlation between different groups of categorical variables

Returns:

* F-test score: Variable between sample group means divided by variation in sample
* p-value: confidence degree

F Test calculates ratio of variation between groups mean over the variatio within each sample group means.

Small F imply poor correlation between variable categories and target variable.
Large F imply strong corrleation between variable categories and target variable.

````python
# ANOVA between "honda" and "subaru"

df_anova = df[['make', 'price']]
grouped_anova = df_anova.groupby(['make'])

anova_results_1 = scipy.stats.f_oneway(grouped_anova.get_group("honda")["price"], grouped_anova.get_group("subaru")["price"])
anova_results_1
````

statistics (F test) < 1, and p-value > 0.05, therefore prices between Honda and Suburas not significantly different.

Let's try honda and jaguar:

````python
anova_results_1 = scipy.stats.f_oneway(grouped_anova.get_group("honda")["price"], grouped_anova.get_group("jaguar")["price"])
anova_results_1
````

Very large F, and p > 0.05  -strong corrleation between a categorical value and other variables.

# Corrleation Statistics

Strength of correlation etween two features, returns:

* Correlation coefficient
```
+1 - large positive
-1 - large negative
0 - no correlation
```

* p-value
Of result:
```
< 0.001 strong certainity
<0.05 moderate certainity
<0.1 weak certainity
> 0.1 No certainity
```

A strong correlation when:
* correlation close to +/-1
* P value < 0.001

````python
# Compare horsepower and price:

pearson_coef, p_value = scipy.stats.pearsonr(df['highway-mpg'], df['price'])
pearson_coef, p_value
````

# Correlation heatmap

Indicates correlation of all variables with one another using Pearson correlation.
Code from: https://ajh1143.github.io/Corr/ and https://seaborn.pydata.org/examples/many_pairwise_correlations.html

````python
# Regular correlation heatmap

# Create Correlation df
corr = df.corr()

fig, ax = plt.subplots(figsize=(10, 10))
colormap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f")

plt.xticks(range(len(corr.columns)), corr.columns);
plt.yticks(range(len(corr.columns)), corr.columns)
plt.show()
````

````python
# Make a nice half one:

corr = df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

fig, ax = plt.subplots(figsize=(10, 10))
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
            
plt.show()
````
